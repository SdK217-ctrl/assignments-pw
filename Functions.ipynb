```python
#3
def square_list(lst):
    return [x*x for x in lst]
```

```python
#4
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
primes = [n for n in range(1, 201) if is_prime(n)]
```

```python
#8
data = [(1, 3), (2, 1), (4, 2)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
#9
celsius = [0, 10, 20]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
#10
s = "hello world"
no_vowels = ''.join(filter(lambda c: c.lower() not in "aeiou", s))
print(no_vowels)
```

```python
#11
orders = [
    [34587, "Learning Python", 4, 40.95],
    [98762, "Python Cookbook", 3, 17.85],
    [77226, "Data Science Handbook", 1, 39.95],
    [88112, "Automate Boring Stuff", 1, 29.95]
]

result = list(map(lambda order: (
    order[0],
    order[2]*order[3] if order[2]*order[3] >= 100 else order[2]*order[3]+10
), orders))
print(result)
```
